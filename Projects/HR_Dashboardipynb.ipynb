```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import count, avg, year, max, col
```

```python
spark = SparkSession.builder \
    .appName("HR Dashboard") \
    .getOrCreate()
```

```python
employee_data = [
    (101, "Ravi", "Male", 25, "IT", "Hyderabad", 50000, "2020-05-10"),
    (102, "Priya", "Female", 28, "HR", "Vijayawada", 45000, "2019-03-15"),
    (103, "Arjun", "Male", 30, "Finance", "Hyderabad", 60000, "2021-07-20"),
    (104, "Sneha", "Female", 26, "IT", "Guntur", 55000, "2022-01-12"),
    (105, "Kiran", "Male", 35, "HR", "Vijayawada", 70000, "2018-11-25"),
    (106, "Divya", "Female", 29, "Finance", "Hyderabad", 65000, "2020-08-18"),
    (107, "Rohit", "Male", 32, "IT", "Guntur", 80000, "2017-06-30"),
    (108, "Anjali", "Female", 27, "HR", "Hyderabad", 48000, "2023-02-14")
]

columns = [
    "EmpID",
    "Name",
    "Gender",
    "Age",
    "Department",
    "City",
    "Salary",
    "HireDate"
]

df = spark.createDataFrame(employee_data, columns)

df.show()
```

```python
gender_count = df.groupBy("Gender").agg(
    count("*").alias("Employee_Count")
)

gender_count.show()
```

```python
avg_age = df.groupBy("Department").agg(
    avg("Age").alias("Average_Age")
)

avg_age.show()
```

```python
hiring_trends = df.withColumn(
    "HireYear",
    year("HireDate")
).groupBy(
    "HireYear"
).count().orderBy(
    "HireYear"
)

hiring_trends.show()
```

```python
city_distribution = df.groupBy("City").count()

city_distribution.show()
```

```python
highest_salary = df.agg(
    max("Salary").alias("MaxSalary")
).collect()[0]["MaxSalary"]

highest_paid = df.filter(
    col("Salary") == highest_salary
)

highest_paid.show()
```

```python
spark.stop()
```
