```python
sales_df = spark.read.csv(
    "sales.csv",
    header=True,
    inferSchema=True
)

sales_df.show(5)
sales_df.printSchema()
```

##Exercise 1: Read sales data → clean nulls → calculate daily sales


```python
df.createOrReplaceTempView("customers")
```

```python
sales_df.createOrReplaceTempView("sales")
```

```python
sales_clean = sales_df.dropna()
```

```python
sales_clean.createOrReplaceTempView("sales_clean")
```

###SQL Query


```python
spark.sql("""
SELECT sale_date,
       SUM(total_amount) AS daily_sales
FROM sales_clean
GROUP BY sale_date
ORDER BY sale_date
""").show()
```

###PySpark

```python
from pyspark.sql.functions import sum

daily_sales = sales_clean.groupBy("sale_date") \
    .agg(sum("total_amount").alias("daily_sales")) \
    .orderBy("sale_date")

daily_sales.show()
```

##Exercise 2: Read customer data → clean invalid rows → city-wise revenue

```python
customers_clean = df.dropna()
```

```python
customers_clean.createOrReplaceTempView("customers_clean")
sales_clean.createOrReplaceTempView("sales_clean")
```

##SQL Query

```python
spark.sql("""
SELECT c.city,
       SUM(s.total_amount) AS city_revenue
FROM customers_clean c
JOIN sales_clean s
ON c.customer_id = s.customer_id
GROUP BY c.city
ORDER BY city_revenue DESC
""").show()
```

###PySpark

```python
from pyspark.sql.functions import sum

joined_df = customers_clean.join(
    sales_clean,
    "customer_id"
)

city_revenue = joined_df.groupBy("city") \
    .agg(sum("total_amount").alias("city_revenue")) \
    .orderBy("city_revenue", ascending=False)

city_revenue.show()
```

###Exercise 3: Find Repeat Customers (>2 Orders)

###SQL


```python
spark.sql("""
SELECT customer_id,
       COUNT(sale_id) AS order_count
FROM sales_clean
GROUP BY customer_id
HAVING COUNT(sale_id) > 2
ORDER BY order_count DESC
""").show()
```

###PySpark

```python
from pyspark.sql.functions import count

repeat_customers = sales_clean.groupBy("customer_id") \
    .agg(count("sale_id").alias("order_count")) \
    .filter("order_count > 2") \
    .orderBy("order_count", ascending=False)

repeat_customers.show()
```

##Exercise 4: Find Highest Spending Customer in Each City

SQL

```python
spark.sql("""
WITH customer_spending AS (
    SELECT
        c.city,
        s.customer_id,
        SUM(s.total_amount) AS total_spend,
        RANK() OVER (
            PARTITION BY c.city
            ORDER BY SUM(s.total_amount) DESC
        ) AS rnk
    FROM customers_clean c
    JOIN sales_clean s
    ON c.customer_id = s.customer_id
    GROUP BY c.city, s.customer_id
)

SELECT city,
       customer_id,
       total_spend
FROM customer_spending
WHERE rnk = 1;
""").show()
```

#PySpark

```python
from pyspark.sql.functions import sum, rank
from pyspark.sql.window import Window

# Join customers and sales
joined_df = customers_clean.join(sales_clean, "customer_id")

# Calculate total spending per customer in each city
customer_spending = joined_df.groupBy("city", "customer_id") \
    .agg(sum("total_amount").alias("total_spend"))

# Rank customers within each city
window_spec = Window.partitionBy("city").orderBy(customer_spending.total_spend.desc())

highest_spending = customer_spending.withColumn(
    "rank",
    rank().over(window_spec)
)

# Keep only the top spender in each city
highest_spending.filter("rank = 1") \
    .select("city", "customer_id", "total_spend") \
    .show()
```

##Exercise 5: Build Final Reporting Table

SQl

```python
spark.sql("""
SELECT
    s.customer_id,
    c.city,
    SUM(s.total_amount) AS total_spend,
    COUNT(s.sale_id) AS order_count
FROM sales_clean s
JOIN customers_clean c
ON s.customer_id = c.customer_id
GROUP BY s.customer_id, c.city
ORDER BY total_spend DESC;
""").show()
```

PySpark

```python
from pyspark.sql.functions import sum, count

final_report = joined_df.groupBy(
    "customer_id",
    "city"
).agg(
    sum("total_amount").alias("total_spend"),
    count("sale_id").alias("order_count")
).orderBy("total_spend", ascending=False)

final_report.show()
```
