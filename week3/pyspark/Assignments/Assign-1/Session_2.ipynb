### WITHCOLUMN

##1.Create bonus column as 10% of salary

```python
df.withColumn("bonus", col("salary") * 0.10).display()
```

##2. Create annual_salary column

```python
df.withColumn("annual_salary", col("salary") * 12).display()
```

##3. Create tax column as 5% of salary

```python
df.withColumn("tax", col("salary") * 0.05).display()
```

##4. Create updated_salary by adding 5000

```python
df.withColumn("updated_salary", col("salary") + 5000).display()
```

##5. Create salary_category column

```python
df.withColumn(
    "salary_category",
    when(col("salary") >= 80000, "High")
    .when(col("salary") >= 50000, "Medium")
    .otherwise("Low")
).display()
```

##6. Create age_group column

```python
df.withColumn(
    "age_group",
    when(col("age") < 25, "Young")
    .otherwise("Adult")
).display()
```

##7.   Create location column by combining city and department

```python
df.withColumn(
    "location",
    concat_ws("-", col("city"), col("department"))
).display()
```

##8. Create increment_salary column with 15% hike

```python
df.withColumn(
    "increment_salary",
    col("salary") * 1.15
).display()
```

##9. Create experience_status column based on joining year

```python
df.withColumn(
    "experience_status",
    when(year("joining_date") < 2021, "Experienced")
    .otherwise("New")
).display()
```

##10.Create name_length column using employee name

```python
df.withColumn(
    "name_length",
    length(col("emp_name"))
).display()
```

##11. Create is_high_salary column using condition

```python
df.withColumn(
    "is_high_salary",
    when(col("salary") > 80000, True)
    .otherwise(False)
).display()
```

##12. Create joining_year column from joining_date

```python
df.withColumn(
    "joining_year",
    year("joining_date")
).display()
```

##13. Create salary_after_tax column

```python
df.withColumn(
    "salary_after_tax",
    col("salary") - (col("salary") * 0.05)
).display()
```

##14. Create department_code column

```python
df.withColumn(
    "department_code",
    substring(col("department"), 1, 3)
).display()
```

##15. Create double_salary column

```python
df.withColumn(
    "double_salary",
    col("salary") * 2
).display()
```
